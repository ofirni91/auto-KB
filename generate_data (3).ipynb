```python
import bs4 as bs
import nltk
import re


def clean_text(text):
    #     text = re.sub(r'\[[0-9]*\]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.lower()
    text = re.sub(r'[@#\$%&\*\(\)\<\>\?\":;\[-]', ' ', text)
    text = re.sub(r'\d', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text

def get_sentence_before_curr(word, filename, beg, end):
    original_beg = beg
    original_end = end
    for year in range(2012, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013/data/' + str(
            year) + '/eval/source_documents/' + filename +'.txt'
        try:
            counter = 0
            with open(PATH, 'r') as f:
                content = f.read()
                while counter < 2:
#                     print(counter)
                    while not content[beg:end].endswith('. ') and not content[beg:end].endswith('.\n') and '</post>' not in content[beg:end]:                    
#                         print(content[beg:end])
#                         print('**************')
                        if 'TEXT>' in content[beg:end]:
                            break
                        end += 1
#                         print(content[beg:end])
                    end += 1
                    counter += 1

                while not content[beg:end].startswith('. ') and not content[beg:end].startswith('.\n') and '</post>' not in content[beg:end]:                    
                    if 'TEXT>' in content[beg:end]:
                        break
                    beg -= 1

                sentences = content[beg:end]
#                 sentences = re.sub('<', '', sentences)
#                 sentences = re.sub('\n', '', sentences)
#                 sentences = re.sub('P>', '', sentences)
#                 sentences = re.sub('<P>', '', sentences)
#                 sentences = re.sub('</P>', '', sentences)
#                 sentences = re.sub('<post>', '', sentences)
#                 sentences = re.sub('</post>', '', sentences)
#                 sentences = re.sub('<TEXT>', '', sentences)
#                 sentences = re.sub('TEXT>', '', sentences)
#                 sentences = re.sub('</TEXT>', '', sentences)
#                 sentences = re.sub('<POSTDATE>', '', sentences)
#                 sentences = re.sub('</POSTDATE>', '', sentences)
#                 sentences = re.sub('</POSTER>', '', sentences)
#                 print('original beg: ', original_beg, 'beg: ', original_beg - beg)
#                 print('original end: ', original_end, 'end: ', original_end - end)
#                 print(sentences)
                return sentences, original_beg - beg, original_end - end
        except:
            pass
    return None, None, None

def get_sentence_new(word, filename, beg, end):
    original_beg = beg
    original_end = end
    for year in range(2012, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013/data/' + str(
            year) + '/eval/source_documents/' + filename +'.txt'
        try:
            with open(PATH, 'r') as f:
                content = f.read()
                while not content[beg:end].endswith('. ') and not content[beg:end].endswith('.\n') and '</post>' not in content[beg:end]:
                    if 'TEXT>' in content[beg:end]:
                        break
                    end += 1
                end += 1
                while not content[beg:end].startswith('. ') and not content[beg:end].startswith('.\n') and '</post>' not in content[beg:end]:
                    if 'TEXT>' in content[beg:end]:
                        break
                    beg -= 1
                end -= 1
                sentences = content[beg:end]
                sentences = re.sub('<', '', sentences)
                sentences = re.sub('\n', '', sentences)
                sentences = re.sub('P>', '', sentences)
                sentences = re.sub('<P>', '', sentences)
                sentences = re.sub('</P>', '', sentences)
                sentences = re.sub('<post>', '', sentences)
                sentences = re.sub('</post>', '', sentences)
                sentences = re.sub('<TEXT>', '', sentences)
                sentences = re.sub('TEXT>', '', sentences)
                sentences = re.sub('</TEXT>', '', sentences)
                sentences = re.sub('<POSTDATE>', '', sentences)
                sentences = re.sub('</POSTDATE>', '', sentences)
                sentences = re.sub('</POSTER>', '', sentences)
                return sentences, original_beg - beg, original_end - end
        except:
            pass
    return None, None, None

def get_sentence(word, filename):
    for year in range(2009, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013 2/data/' + str(
            year) + '/eval/source_documents/'
        text = ""
        try:
            source = open(PATH + filename + '.txt').read()
            soup = bs.BeautifulSoup(source, 'html.parser')
            for paragraph in soup.find_all('p'):
                text += paragraph.text

            text = re.sub(r'\n', ' ', text)
            sentences = nltk.sent_tokenize(text)
            result = list()
            for sentence in sentences:
#                 print(sentence)
                sent_no_spaces = clean_text(sentence)
                sent_no_spaces = re.sub(r' ', '', sent_no_spaces)
                word_no_spaces = clean_text(word)
                word_no_spaces = re.sub(r' ', '', word_no_spaces)
                if word_no_spaces in sent_no_spaces:
                    result.append(clean_text(sentence))

            return result
        except:
            print('trying next year!')

            
def get_text(word, filename):
    for year in range(2012, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013 2/data/' + str(
            year) + '/eval/source_documents/'
        text = ""
        try:
            source = open(PATH + filename + '.txt').read()
            for word in source:
                text += word
            return text
        except:
            print('trying next year!')

def get_paragraph(filename):
    for year in range(2009, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013 2/data/' + str(
            year) + '/eval/source_documents/'
        text = ""
        try:
            source = open(PATH + filename + '.txt').read()
            soup = bs.BeautifulSoup(source, 'html.parser')
            for paragraph in soup.find_all('p'):
                text += paragraph.text

            return clean_text(text)
        except:
            print('trying next year!')
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np
import itertools
import spacy
from IPython.display import clear_output

# nlp = spacy.load('en_core_web_lg')
# nlp = spacy.load('en_core_web_sm')
all_years_datasets = list()
total_permutations = 0
count_empty_sent = 0
for year in range(2012, 2014):
    file_name = '/Users/onijinsky/PycharmProjects/unlinkable_entities/tac_kbp_' + str(
        year) + '_english_entity_linking_evaluation_KB_links.tab'
    tab_file_df = pd.read_csv(file_name, sep='\t', header=None)
    tab_file_df = tab_file_df.iloc[:, 0:3]
    tab_file_df.columns = ['query', 'entity_id', 'type']

    path = '/Users/onijinsky/PycharmProjects/unlinkable_entities/tac_kbp_' + str(
        year) + '_english_entity_linking_evaluation_queries.xml'
    root = ET.parse(path).getroot()
    names = list()
    docs = list()
    queries = list()
    begs = list()
    ends = list()
    for child in root:
        queries.append(child.attrib['id'])
        for c in child:
            if c.tag == 'name':
                names.append(c.text)
            if c.tag == 'docid':
                docs.append(c.text)
            if c.tag == 'beg':
                begs.append(c.text)
            if c.tag == 'end':
                ends.append(c.text)

    data_without_entitiy_id = pd.DataFrame(
        {'query': queries,
         'name': names,
         'docid': docs,
         'beg': begs,
         'end': ends
         })

    df = pd.merge(tab_file_df, data_without_entitiy_id, on='query')

    text_a = list()
    text_b = list()
    mention_a = list()
    mention_b = list()
    entity_id_a = list()
    entity_id_b = list()
    same_entity = list()
    db_entity_a = list()
    db_entity_b = list()
    a_vec = list()
    b_vec = list()
    a_sen_vec = list()
    b_sen_vec = list()
    a_sen = list()
    b_sen = list()
    a_text = list()
    b_text = list()
    a_beg_end = list()
    b_beg_end = list()
    
    for name, group in df.groupby(['entity_id']):
        #         group['name'] = group['name'].apply(lambda x: x.lower())
#         group = group.drop_duplicates(subset='name', keep='first')
        subsets = itertools.combinations(list(group['query']), 2)
        for s in subsets:
            a = group[group['query'] == s[0]]
            b = group[group['query'] == s[1]]
            if a.name.item() != b.name.item():
                continue
            # a
            text_a.append(a.docid.item())
            mention_a.append(a.name.item())
            entity_id_a.append(a.entity_id.item())
            mention_a_vec = embed([a.name.item()])
            a_vec.append(mention_a_vec)
            a_text.append(get_text(a.name.item(), a.docid.item()))
            a_beg_end.append((int(a.beg.item()), int(a.end.item())))
            sen_lst_a = [get_sentence_new(a.name.item(), a.docid.item(), int(a.beg.item()), int(a.end.item()))]
            if len(sen_lst_a) > 0:
                sen_sim = 0.0
                for sen_a in sen_lst_a:
                    sen_sim += embed([sen_a])['outputs']
                avg_sen_sim_a = sen_sim / len(sen_lst_a)
            else:
                avg_sen_sim_a = np.array([0] * 512)
                print(a.name.item())
                print(a.docid.item())
                count_empty_sent += 1
                print('**********')
            a_sen_vec.append(avg_sen_sim_a)
            a_sen.append(sen_a)
            # TODO: add paragraph similarity for a and b!!!

            # b
            text_b.append(b.docid.item())
            mention_b.append(b.name.item())
            entity_id_b.append(b.entity_id.item())
            mention_b_vec = embed([b.name.item()])
            b_vec.append(mention_b_vec)
            b_text.append(get_text(b.name.item(), b.docid.item()))
            b_beg_end.append((int(b.beg.item()), int(b.end.item())))
            sen_lst_b = [get_sentence_new(b.name.item(), b.docid.item(), int(b.beg.item()), int(b.end.item()))]
            if len(sen_lst_b) > 0:
                sen_sim = 0.0
                for sen_b in sen_lst_b:
                    sen_sim += embed([sen_b])['outputs']
                avg_sen_sim_b = sen_sim / len(sen_lst_a)
            else:
                avg_sen_sim_b = np.array([0] * 512)
                print(b.name.item())
                print(b.docid.item())
                count_empty_sent += 1
                print('**********')
            b_sen_vec.append(avg_sen_sim_b)
            b_sen.append(sen_b)
            
            if a.entity_id.item() == b.entity_id.item():
                same_entity.append(1)
            else:
                same_entity.append(0)

            db_entity_a.append(a.entity_id.item())
            db_entity_b.append(b.entity_id.item())

    positive_dataset = pd.DataFrame({'text_a': text_a,
                                     'text_b': text_b,
                                     'mention_a': mention_a,
                                     'mention_b': mention_b,
                                     'entity_id_a': entity_id_a,
                                     'entity_id_b': entity_id_b,
                                     'same_entity': same_entity,
                                     'kb_entity_a': db_entity_a,
                                     'kb_entity_b': db_entity_b,
                                     'a_word_vec': a_vec,
                                     'b_word_vec': b_vec,
                                     'a_sen_vec': a_sen_vec,
                                     'b_sen_vec': b_sen_vec,
                                     'a_sen': a_sen,
                                     'b_sen': b_sen,
                                     'all_text_a': a_text,
                                     'a_word_loc': a_beg_end,
                                     'all_text_b': b_text,
                                     'b_word_loc': b_beg_end
                                     })
    all_years_datasets.append(positive_dataset)

final_dataset = pd.concat(all_years_datasets, ignore_index=True)
final_dataset.to_csv('tac_positive_pairs_trivials_google_3.csv')
```

```python
def get_locs(string_locs):
    import re
    locations = string_locs
    locations = re.sub('\[', '', locations)
    locations = re.sub('\]', '', locations)
    locations = re.sub('\(', '', locations)
    locations = re.sub('\)', '', locations)
    locations = re.sub('\s+', '', locations)
    int_locs = [int(x) for x in locations.split(',') if len(x) > 0]
    tuple_locs = []
    for i in range(0, len(int_locs), 2):
        tuple_locs.append((int_locs[i], int_locs[i + 1]))
    return tuple_locs

def get_loc(row, pair):
    str_loc = row[pair + '_word_loc']
    str_loc = re.sub('\(', '', str_loc)
    str_loc = re.sub('\)', '', str_loc)
    loc = str_loc.split(',')
    return int(loc[0]), int(loc[1])
```

```python
def get_sentence_new(content, beg, end):
    original_beg = beg
    original_end = end    
    while not content[beg:end].endswith('.') and end != len(content):
        end += 1
    while not content[beg:end].startswith('.') and beg != 0:
        beg -= 1
    sentences = content[beg:end]
    return sentences, original_beg - beg, original_end - end
```

```python
def get_sentence_before_curr(word, filename, beg, end):
    original_beg = beg
    original_end = end
    counter = 0
    content = filename
    while counter < 2:
        while (not content[beg:end].endswith('.') and not content[beg:end].endswith('. ')) and end < len(content):
            end += 1
        end += 1
        counter += 1
    while (not content[beg:end].startswith('.') and not content[beg:end].startswith('\"')) and beg > 0:
        beg -= 1
    sentences = content[beg:end]
    return sentences, original_beg - beg, original_end - end
```
