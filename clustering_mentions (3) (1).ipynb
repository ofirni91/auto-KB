```python
def str_to_vec(str_word_vec):
    str_word_vec = str_word_vec[12:-33]
    str_word_vec = re.sub(r'\[', '', str_word_vec)
    str_word_vec = re.sub(r'\]', '', str_word_vec)
    str_word_vec = re.sub(r'\n', ' ', str_word_vec)
    str_word_vec = re.sub(r'\s+', ' ', str_word_vec)
    vec = np.fromstring(str_word_vec, dtype=float, sep=' ')
    return vec

def str_to_vec_rep(str_word_vec):
    str_word_vec = str_word_vec[79:-25]
    str_word_vec = re.sub(r'\[', '', str_word_vec)
    str_word_vec = re.sub(r'\]', '', str_word_vec)
    str_word_vec = re.sub(r'\n', ' ', str_word_vec)
    str_word_vec = re.sub(r'\s+', ' ', str_word_vec)
    vec = np.fromstring(str_word_vec, dtype=float, sep=',')
    return vec

def str_to_vec_yago(str_word_vec):
    str_word_vec = str_word_vec[12:-33]
    str_word_vec = re.sub(r'\[', '', str_word_vec)
    str_word_vec = re.sub(r'\]', '', str_word_vec)
    str_word_vec = re.sub(r'\n', ' ', str_word_vec)
    str_word_vec = re.sub(r'\s+', ' ', str_word_vec)
    vec = np.fromstring(str_word_vec, dtype=float, sep=' ')
    return vec


def str_to_vec_rep_yago(str_word_vec):
    str_word_vec = str_word_vec[79:-25]
    str_word_vec = re.sub(r'\[', '', str_word_vec)
    str_word_vec = re.sub(r'\]', '', str_word_vec)
    str_word_vec = re.sub(r'\n', ' ', str_word_vec)
    str_word_vec = re.sub(r'\s+', ' ', str_word_vec)
    vec = np.fromstring(str_word_vec, dtype=float, sep=',')
    return vec
```

```python
def get_locs(string_locs):
    import re
    locations = string_locs
    locations = re.sub('\[', '', locations)
    locations = re.sub('\]', '', locations)
    locations = re.sub('\(', '', locations)
    locations = re.sub('\)', '', locations)
    locations = re.sub('\s+', '', locations)
    int_locs = [int(x) for x in locations.split(',') if len(x) > 0]
    tuple_locs = []
    for i in range(0, len(int_locs), 2):
        tuple_locs.append((int_locs[i], int_locs[i + 1]))
    return tuple_locs
```

```python
def get_sentence_before_curr(word, filename, beg, end):
    original_beg = beg
    original_end = end
    counter = 0
    content = filename
    while counter < 2:
        while (not content[beg:end].endswith('.') and not content[beg:end].endswith('. ')) and end < len(content):
            end += 1
        end += 1
        counter += 1
    while (not content[beg:end].startswith('.') and not content[beg:end].startswith('\"')) and beg > 0:
        beg -= 1
    sentences = content[beg:end]
    return sentences, original_beg - beg, original_end - end
```

```python
import bs4 as bs
import nltk
import re

def get_loc(str_loc):
    str_loc = re.sub('\(', '', str_loc)
    str_loc = re.sub('\)', '', str_loc)
    loc = str_loc.split(',')
    return int(loc[0]), int(loc[1])

def get_locs(string_locs):
    import re
    locations = string_locs
    locations = re.sub('\[', '', locations)
    locations = re.sub('\]', '', locations)
    locations = re.sub('\(', '', locations)
    locations = re.sub('\)', '', locations)
    locations = re.sub('\s+', '', locations)
    int_locs = [int(x) for x in locations.split(',') if len(x) > 0]
    tuple_locs = []
    for i in range(0, len(int_locs), 2):
        tuple_locs.append((int_locs[i], int_locs[i + 1]))
    return tuple_locs

def clean_text(text):
    #     text = re.sub(r'\[[0-9]*\]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.lower()
    text = re.sub(r'[@#\$%&\*\(\)\<\>\?\":;\[-]', ' ', text)
    text = re.sub(r'\d', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text

def get_sentence_before_curr(word, filename, beg, end):
    original_beg = beg
    original_end = end
    for year in range(2012, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013/data/' + str(
            year) + '/eval/source_documents/' + filename +'.txt'
        try:
            counter = 0
            with open(PATH, 'r') as f:
                content = f.read()
                while counter < 2:
#                     print(counter)
                    while not content[beg:end].endswith('. ') and not content[beg:end].endswith('.\n') and '</post>' not in content[beg:end]:                    
#                         print(content[beg:end])
#                         print('**************')
                        if 'TEXT>' in content[beg:end]:
                            break
                        end += 1
#                         print(content[beg:end])
                    end += 1
                    counter += 1

                while not content[beg:end].startswith('. ') and not content[beg:end].startswith('.\n') and '</post>' not in content[beg:end]:                    
                    if 'TEXT>' in content[beg:end]:
                        break
                    beg -= 1

                sentences = content[beg:end]
#                 sentences = re.sub('<', '', sentences)
#                 sentences = re.sub('\n', '', sentences)
#                 sentences = re.sub('P>', '', sentences)
#                 sentences = re.sub('<P>', '', sentences)
#                 sentences = re.sub('</P>', '', sentences)
#                 sentences = re.sub('<post>', '', sentences)
#                 sentences = re.sub('</post>', '', sentences)
#                 sentences = re.sub('<TEXT>', '', sentences)
#                 sentences = re.sub('TEXT>', '', sentences)
#                 sentences = re.sub('</TEXT>', '', sentences)
#                 sentences = re.sub('<POSTDATE>', '', sentences)
#                 sentences = re.sub('</POSTDATE>', '', sentences)
#                 sentences = re.sub('</POSTER>', '', sentences)
#                 print('original beg: ', original_beg, 'beg: ', original_beg - beg)
#                 print('original end: ', original_end, 'end: ', original_end - end)
#                 print(sentences)
                return sentences, original_beg - beg, original_end - end
        except:
            pass
    return None, None, None

def get_sentence_new(word, filename, beg, end):
    original_beg = beg
    original_end = end
    for year in range(2012, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013/data/' + str(
            year) + '/eval/source_documents/' + filename +'.txt'
        try:
            with open(PATH, 'r') as f:
                content = f.read()
                while not content[beg:end].endswith('. ') and not content[beg:end].endswith('.\n') and '</post>' not in content[beg:end]:
                    if 'TEXT>' in content[beg:end]:
                        break
                    end += 1
                end += 1
                while not content[beg:end].startswith('. ') and not content[beg:end].startswith('.\n') and '</post>' not in content[beg:end]:
                    if 'TEXT>' in content[beg:end]:
                        break
                    beg -= 1
                end -= 1
                sentences = content[beg:end]
                sentences = re.sub('<', '', sentences)
                sentences = re.sub('\n', '', sentences)
                sentences = re.sub('P>', '', sentences)
                sentences = re.sub('<P>', '', sentences)
                sentences = re.sub('</P>', '', sentences)
                sentences = re.sub('<post>', '', sentences)
                sentences = re.sub('</post>', '', sentences)
                sentences = re.sub('<TEXT>', '', sentences)
                sentences = re.sub('TEXT>', '', sentences)
                sentences = re.sub('</TEXT>', '', sentences)
                sentences = re.sub('<POSTDATE>', '', sentences)
                sentences = re.sub('</POSTDATE>', '', sentences)
                sentences = re.sub('</POSTER>', '', sentences)
                return sentences, original_beg - beg, original_end - end
        except:
            pass
    return None, None, None

def get_sentence(word, filename):
    for year in range(2009, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013 2/data/' + str(
            year) + '/eval/source_documents/'
        text = ""
        try:
            source = open(PATH + filename + '.txt').read()
            soup = bs.BeautifulSoup(source, 'html.parser')
            for paragraph in soup.find_all('p'):
                text += paragraph.text

            text = re.sub(r'\n', ' ', text)
            sentences = nltk.sent_tokenize(text)
            result = list()
            for sentence in sentences:
#                 print(sentence)
                sent_no_spaces = clean_text(sentence)
                sent_no_spaces = re.sub(r' ', '', sent_no_spaces)
                word_no_spaces = clean_text(word)
                word_no_spaces = re.sub(r' ', '', word_no_spaces)
                if word_no_spaces in sent_no_spaces:
                    result.append(clean_text(sentence))

            return result
        except:
            print('trying next year!')

            
def get_text(word, filename):
    for year in range(2012, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013 2/data/' + str(
            year) + '/eval/source_documents/'
        text = ""
        try:
            source = open(PATH + filename + '.txt').read()
            for word in source:
                text += word
            return text
        except:
            print('trying next year!')

def get_paragraph(filename):
    for year in range(2009, 2014):
        PATH = '/Users/onijinsky/Downloads/tac_kbp_eng_ent_link_comp_2009-2013 2/data/' + str(
            year) + '/eval/source_documents/'
        text = ""
        try:
            source = open(PATH + filename + '.txt').read()
            soup = bs.BeautifulSoup(source, 'html.parser')
            for paragraph in soup.find_all('p'):
                text += paragraph.text

            return clean_text(text)
        except:
            print('trying next year!')
```
